```python
from ops_capstone import Curves

galaxies= ['ddo154','ddo170','ngc1560','ngc2403','ngc2841',
           'ngc2903','ngc3109','ngc3198','ngc6503','ngc7331','ugc2259']

for galaxy in galaxies:

    curve = Curves()

    print(galaxy.upper())
    #curve.add_model(model_name='sem ME') #model without DM
    curve.add_model(model_name='NFW') #model NFW
    curve.add_model(model_params=[[0,2,2]]) #model classical modified
    curve.add_model(model_params=[[1,1,2]]) #model alternative
    
    
    

    curve.fit_galaxy(galaxy, figsize=(10,10))
```
